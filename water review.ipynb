```python
#https://stackoverflow.com/questions/31328861/python-pandas-replacing-header-with-top-row
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
```

```python
# https://www.ritchieng.com/pandas-selecting-multiple-rows-and-columns/
sfusd = df[df.DISTRICT=='San Francisco Unified']
```

```python
sfusd.head()
```

```python
# how many fountains/etc test at or under the 5 ug/L threshold? 
#sfusd.groupby('XMOD')['RESULT'].value_counts()
sfusd.groupby(["XMOD", "RESULT"]).size()
```

```python
# how many fountains/etc test above the 5 ug/L threshold?
sfusd[sfusd.RESULT > 5].SchoolName.describe()
```

```python
#which schools had fountains that tested above the threshold?
sfusd[sfusd.RESULT > 5].SchoolName
```

```python
#which schools does the above correspond to?
sfusd.groupby('RESULT')['SchoolName'].value_counts()
```

```python
#get more specifics on each of the counts for schools with results higher than 5
sfusd.groupby('SchoolName')['RESULT'].value_counts()
```

```python
# https://stackoverflow.com/questions/24988873/python-sort-descending-dataframe-with-pandas
# https://medium.com/@msalmon00/helpful-python-code-snippets-for-data-exploration-in-pandas-b7c5aed5ecb9
sfusd_sort = sfusd.sort_values(['RESULT','XMOD','SchoolName'], ascending=[0,1,1]) 
print(sfusd_sort.head())
```

```python
# look for specific school results
# https://stackoverflow.com/questions/26640129/search-for-string-in-all-pandas-dataframe-columns-and-filter
CIS_mask = np.column_stack([sfusd[col].str.contains("Chinese Immersion School at DeAvila", na=False) for col in sfusd])
sfusd.loc[CIS_mask.any(axis=1)]
```
